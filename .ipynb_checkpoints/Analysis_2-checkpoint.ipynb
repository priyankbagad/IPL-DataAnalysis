```python
#most destructive death overs batsman
#based on strike rate
#criteria : minimum 200 balls in 16-20 overs
```

```python
death_over=data[data['over']>15]
```

```python
all_batsmen = death_over.groupby('batsman')['batsman_runs'].count()
```

```python
x = all_batsmen>200
```

```python
all_batsmen[x].shape
```

```python
batsman_list=all_batsmen[x].index.tolist() #who played 200 balls
```

```python
final = data[data['batsman'].isin(batsman_list)]
```

```python
run = final.groupby('batsman')['batsman_runs'].sum()
```

```python
balls = final.groupby('batsman')['batsman_runs'].count()
```

```python
sr = (run/balls)*100
```

```python
sr
```

```python
#which team score more in which over using pivot table
```

```python
mask = data['batsman_runs']==6
six = data[mask]
six.head()
```

```python
pt=six.pivot_table(index='over',columns='batting_team',values='batsman_runs',aggfunc='count')
```

```python
import seaborn as sns
```

```python
sns.heatmap(pt) #death overs more sixes and 6th over more sixes because of powerplay
```
